### 1. import numpy, create a random integer vector of length 15

```python
import numpy as np
a = np.random.randint(5, size=15)
print(a)
```

### 2. get the 3 largest values in the array

```python
print(a[np.argsort(a)[-3:]]) #print (a[np.argpartition(-a,3)[:3]])
```

### 3. find the most frequent value in this array

```python
print(np.bincount(a).argmax())
```

### 4. shuffle it, compute the averages with a sliding window of size 5

```python
def moving_average(a, n=5) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
np.random.shuffle(a)
print(moving_average(a, n=5))
```

### 5. find the nearest value to a given value (z) in this array

```python
z = 2.1
m = a.flat[np.abs(a - z).argmin()]
print(m)
```

### 6.  generate a 2D Gaussian kernel 

```python
X, Y = np.meshgrid(np.linspace(-1,1,10), np.linspace(-1,1,10))
D = np.sqrt(X*X+Y*Y)
sigma, mu = 1.0, 0.0
G = np.exp(-( (D-mu)**2 / ( 2.0 * sigma**2 ) ) )
print(G/np.sum(G))
```

### 7. get a random video from the internet, show the 100th frame

```python
import cv2
import matplotlib.pyplot as plt

# test on high-resolution
plt.figure(dpi=300)

img = cv2.imread('testimg.png')

# test on BGR
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

### 8. blur it by using the generated Gaussian kernel

```python
# note the sum
blur_img = cv2.filter2D(RGB_img, -1, G/np.sum(G))
plt.figure(dpi=300)
plt.imshow(blur_img)
```
